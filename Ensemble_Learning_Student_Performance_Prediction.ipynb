## **Part B: Dataset Understanding & Preparation**

```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/students_dataset.csv')
```

```python
data.columns
```

### Task 6: Feature & Target Identification

```python
X = data.drop(columns=['student_id','age','device_type','course_start_date'])
```

```python
y_cls = data.completion_status
```

```python
y_reg = data.final_score
```

### Tasl 8: Data Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# data before preprocessing
X.head()
```

```python
# Cheacking missing values
X.isna().sum()
```

```python
# Filling missing values with 0
X.fillna(0,inplace=True)
```

```python
# Cheacking Datatypes
X.info()
```

```python
# Making variables for Encoder with and without order

education = ['HighSchool','Undergrad','Graduate','WorkingPro']
course_lvl = ['Beginner','Intermediate','Advanced']
country = X['country_region'].unique()
course_cat = X['course_category'].unique()
```

```python
# Using Ordinal Encoder for Encoding Categorical features
OE = OrdinalEncoder(
    categories=[education,course_lvl,country,course_cat]
)
```

```python
# Replacing values to categorical to numeric
X[['education_background','course_level','country_region','course_category']] = OE.fit_transform(X[['education_background','course_level','country_region','course_category']])
```

```python
# Data Overview after Preprocessing
X.head()
```

### Task 7: Train-Test Split

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Train-Test-Split for Regression 
X_train_reg,X_test_reg,y_train_reg,y_test_reg = train_test_split(X,y_reg,test_size=0.2,random_state=42)
```

```python
# Train-Test-Split for Classification 
X_train_cls,X_test_cls,y_train_cls,y_test_cls = train_test_split(X,y_cls,test_size=0.2,random_state=42)
```

## **Part C: Bagging (Bootstrap Aggregation)**

```python
from sklearn.ensemble import BaggingClassifier,BaggingRegressor
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

### Task: 9: Train a Bagging Classifier

```python
bag_cls = BaggingClassifier(random_state=42)
```

```python
# model training
bag_cls.fit(X_train_cls,y_train_cls)
```

### Task 10: Train a Bagging Regressor

```python
bag_reg = BaggingRegressor(random_state=42)
```

```python
# model training
bag_reg.fit(X_train_reg,y_train_reg)
```

### Task 11: Compare bagging results with a single base model

```python
# creating model with parameters to prevent overfitting
tree_reg = DecisionTreeRegressor(max_depth=5, min_samples_split=2, random_state=42)
tree_cls = DecisionTreeClassifier(max_depth=5, min_samples_split=2, random_state=42)
```

```python
# training regression model
tree_reg.fit(X_train_reg,y_train_reg)
```

```python
# training classification model
tree_cls.fit(X_train_cls,y_train_cls)
```

```python
# Regressor model
bag_pred_reg = bag_reg.predict(X_test_reg)
tree_pred_reg = tree_reg.predict(X_test_reg)
```

```python
# Classification model
bag_pred_cls = bag_cls.predict(X_test_cls)
tree_pred_cls = tree_cls.predict(X_test_cls)
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
pd.DataFrame({
    'R2 Score': [r2_score(y_test_reg,tree_pred_reg),r2_score(y_test_reg,bag_pred_reg)],
    'Accuracy Score': [r2_score(y_test_cls,tree_pred_cls),r2_score(y_test_cls,bag_pred_cls)]
},index=['Decision Tree','Bagging Model'])
```
